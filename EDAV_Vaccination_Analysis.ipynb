```python
'''Question 8 - Examine the impact of using dropna() versus fillna(method='ffill') on cumulative vaccination totals for India and the USA.
Compare summary statistics before and after each approach.
This question is attempted first since it covers preprocessing phase using pandas'''

filt_df = df[df['country'].isin(['India','United States'])]
filt_df = filt_df[['country', 'date', 'total_vaccinations']] #picking only relevant columns for filling null values

print("Actual Summary:\n")
print(filt_df.groupby('country')['total_vaccinations'].describe())
```

```python
# 2. Using dropna()
df_dropna = filt_df.dropna()

print("\nAfter dropna():\n")
print(df_dropna.groupby('country')['total_vaccinations'].describe())
```

```python
# 3. Using forward fill
df_ffill = filt_df.copy()
df_ffill['total_vaccinations'] = df_ffill.groupby('country')['total_vaccinations'].fillna(method='ffill')

print("\nAfter forward fill (ffill):\n")
print(df_ffill.groupby('country')['total_vaccinations'].describe())
```

### Comparison: dropna() vs fillna(method='ffill')

- `dropna()` removes all rows with missing values, reducing the dataset size and potentially losing important time-series information.
- `fillna(method='ffill')` replaces missing values with the previous valid value, preserving continuity in the data.

### Conclusion

For cumulative vaccination data, `ffill()` is generally better because vaccination counts do not decrease over time. Forward filling maintains realistic trends, whereas `dropna()` may discard useful records and break the continuity of the data.

```python
''' Question 2 - Examine the relationship between people_vaccinated and people_fully_vaccinated using ufuncs such as np.log and np.sqrt.
Assess whether full vaccination rates are proportional to initial vaccination rates.'''
import numpy as np

# Selecting relevant columns and drop missing values
df_rel = df[['people_vaccinated', 'people_fully_vaccinated']].dropna()

# Remove zero values to avoid log issues
df_rel = df_rel[(df_rel['people_vaccinated'] > 0) & (df_rel['people_fully_vaccinated'] > 0)]

# Applying ufuncs
df_rel['log_vaccinated'] = np.log(df_rel['people_vaccinated'])
df_rel['log_fully_vaccinated'] = np.log(df_rel['people_fully_vaccinated'])

df_rel['sqrt_vaccinated'] = np.sqrt(df_rel['people_vaccinated'])
df_rel['sqrt_fully_vaccinated'] = np.sqrt(df_rel['people_fully_vaccinated'])
```

```python
# Correlation (original)
print("Correlation (Original):",
      df_rel['people_vaccinated'].corr(df_rel['people_fully_vaccinated']))

# Correlation (log transformed)
print("Correlation (Log):",
      df_rel['log_vaccinated'].corr(df_rel['log_fully_vaccinated']))

# Correlation (sqrt transformed)
print("Correlation (Sqrt):",
      df_rel['sqrt_vaccinated'].corr(df_rel['sqrt_fully_vaccinated']))
```

### Relationship between Vaccinated and Fully Vaccinated

- Strong positive correlation is observed between `people_vaccinated` and `people_fully_vaccinated` in original, log, and sqrt transformations.
- Log and sqrt transformations help stabilize variance and confirm the consistency of the relationship.

### Conclusion

Full vaccination rates are approximately proportional to initial vaccination rates, as indicated by consistently high correlation across transformations.

```python
''' Question - 4: Create a 'Vaccination Efficiency Score' by computing the ratio of fully vaccinated to vaccinated people.
Rank countries and construct a comparative DataFrame using index-aligned operations.'''
# Select required columns and drop missing values
df_eff = df[['country', 'people_vaccinated', 'people_fully_vaccinated']].dropna()

# Group by country (take max since data is cumulative)
vaccinated = df_eff.groupby('country')['people_vaccinated'].max()
fully_vaccinated = df_eff.groupby('country')['people_fully_vaccinated'].max()

# Efficiency Score (index-aligned)
efficiency_score = fully_vaccinated / vaccinated

# Create comparative DataFrame
eff_df = pd.DataFrame({
    'people_vaccinated': vaccinated,
    'people_fully_vaccinated': fully_vaccinated,
    'efficiency_score': efficiency_score
})

# Ranking
eff_df['rank'] = eff_df['efficiency_score'].rank(ascending=False)

# Sort by efficiency
eff_df_sorted = eff_df.sort_values(by='efficiency_score', ascending=False)

# Display top 10
print("Top 10 Countries by Efficiency:\n", eff_df_sorted.head(10))
```

### Vaccination Efficiency Score

- The efficiency score is computed as the ratio of `people_fully_vaccinated` to `people_vaccinated`.
- Index-aligned operations ensure accurate computation across countries.
- Countries are ranked based on efficiency score in descending order.

### Conclusion

Higher efficiency scores indicate better conversion from initial vaccination to full vaccination, reflecting effective vaccination strategies.

```python
''' Question - 5 : Identify columns with more than 30% missing values. Examine whether to drop them or impute.
Apply forward-fill (ffill) for time-series columns and justify this choice with evidence.'''

# First let's identify columns with >30% missing values
missing_percent = df.isnull().mean() * 100
cols_30 = missing_percent[missing_percent > 30]

print("Columns with >30% missing values:\n", cols_30)
```

```python
# Drop columns with very high missing values (example threshold 50%)
cols_to_drop = missing_percent[missing_percent > 50].index

df.drop(columns=cols_to_drop, inplace=True)

print("Dropped columns:\n", cols_to_drop)
```

```python
# Ensure proper sorting for time-series
df['date'] = pd.to_datetime(df['date'])
df.sort_values(['country', 'date'], inplace=True)

# Apply forward fill per country
df.update(df.groupby('country').ffill())

print("After ffill:\n", df.head())
```

```python
print("Missing values before handling:\n", missing_percent)

print("\nMissing values after ffill:\n", df.isnull().mean() * 100)
```

### Missing Value Analysis and Handling

- Columns with more than 30% missing values were identified using percentage calculation.
- Columns with very high missing values (>50%) were dropped due to insufficient data.

- Forward fill (`ffill`) was applied to time-series data grouped by country.
- This method preserves continuity in cumulative vaccination data.

### Justification

Since vaccination data is cumulative and does not decrease over time, forward filling is appropriate as it maintains realistic trends. The reduction in missing values after applying `ffill` supports this choice.

```python
''' Question - 7: Design a daily vaccination rate calculator using apply() and diff().
Combine results from multiple countries into a unified hierarchical DataFrame and highlight top-performing nations.'''

# Ensure proper sorting
df['date'] = pd.to_datetime(df['date'])
df.sort_values(['country', 'date'], inplace=True)

# Daily vaccination rate
df['daily_vaccinations_calc'] = df.groupby('country')['total_vaccinations'].diff()

# Handling negative/NaN values using apply()
df['daily_vaccinations_calc'] = df['daily_vaccinations_calc'].apply(lambda x: x if x >= 0 else 0)

print(df[['country', 'date', 'daily_vaccinations_calc']].head())
```

```python
# Hierarchical index
df_daily = df.set_index(['country', 'date'])
print(df_daily.head())
```

```python
countries = ['India', 'United States', 'Brazil']

df_selected = df_daily.loc[countries]

print(df_selected.head())
```

```python
# Average daily vaccination per country
avg_daily = df.groupby('country')['daily_vaccinations_calc'].mean()

# Top 10 countries
top_countries = avg_daily.sort_values(ascending=False).head(10)

print("Top Performing Countries:\n", top_countries)
```

### Daily Vaccination Rate Analysis

- Daily vaccination rates are computed using `diff()` on cumulative totals.
- `apply()` is used to handle negative or invalid values.
- A hierarchical DataFrame (country, date) enables multi-level analysis.
- Top-performing countries are identified based on average daily vaccinations.

### Conclusion

Using `diff()` provides an efficient way to compute daily changes, while hierarchical indexing allows comparison across countries and time.

```python
''' Question - 10 : Build a multi-country vaccination comparison using pivot_table with hierarchical indexing on country and vaccine type.
Export a structured summary and push the notebook to GitHub.'''

# Create pivot table with hierarchical index
pivot_df = pd.pivot_table(df, values='total_vaccinations',
    index=['country', 'vaccines'],   # hierarchical index
    aggfunc='max'                   # cumulative → take max
)

print(pivot_df.head(10))
```
